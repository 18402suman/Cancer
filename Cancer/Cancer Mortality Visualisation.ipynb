```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Load the dataset
file_path = "Transformed_Cancer_Mortality.csv"
df = pd.read_csv(file_path)

# Ensure data is sorted by year
df = df.sort_values(by=["region", "Year"])

# Unique regions
regions = df["region"].unique()
colors = plt.cm.viridis(np.linspace(0, 1, len(regions)))

# Figure 1: Cumulative Mortality over Time by Region
plt.figure(figsize=(12, 6))
for region, color in zip(regions, colors):
    region_data = df[df["region"] == region]
    plt.fill_between(region_data["Year"], region_data["Cumulative_Deaths"], alpha=0.5, color=color, label=region)
    plt.text(region_data["Year"].max(), region_data["Cumulative_Deaths"].max(), f"{region_data['Cumulative_Deaths'].max():,.0f}", ha='left', fontsize=8, color=color)
plt.title("Cumulative Cancer Mortality Over Time by Region")
plt.xlabel("Year")
plt.ylabel("Cumulative Deaths")
plt.legend(fontsize=6, loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()

# Figure 2: Annual Mortality per Region (Line Plot)
plt.figure(figsize=(12, 6))
for region, color in zip(regions, colors):
    region_data = df[df["region"] == region]
    plt.plot(region_data["Year"], region_data["Deaths"], marker='o', linestyle='-', color=color, label=region)
plt.title("Annual Cancer Mortality Per Region")
plt.xlabel("Year")
plt.ylabel("New Deaths Per Year")
plt.legend(fontsize=6, loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()

# Figure 3: Heatmap of Mortality Growth Rate by Region and Year
pivot_table = df.pivot(index="region", columns="Year", values="Growth_Rate")
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap="coolwarm", annot=True, fmt=".1f")
plt.title("Growth Rate of Cancer Mortality by Region and Year")
plt.xlabel("Year")
plt.ylabel("Region")
plt.show()

# Figure 4: Boxplot of Annual Mortality by Region
plt.figure(figsize=(12, 6))
sns.boxplot(x="region", y="Deaths", data=df, palette="viridis")
plt.xticks(rotation=90)
plt.title("Distribution of Annual Cancer Mortality by Region")
plt.xlabel("Region")
plt.ylabel("Annual Mortality")
plt.show()

# Figure 5: Scatter Plot of Mortality vs Growth Rate
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["Growth_Rate"], y=df["Deaths"], hue=df["region"], palette="viridis", alpha=0.7)
plt.title("Relationship Between Growth Rate and Mortality")
plt.xlabel("Growth Rate (%)")
plt.ylabel("Mortality")
plt.grid(True)
plt.show()

# Figure 6: Rolling Average Mortality per Region
plt.figure(figsize=(12, 6))
for region, color in zip(regions, colors):
    region_data = df[df["region"] == region]
    plt.plot(region_data["Year"], region_data["Rolling_Avg_Deaths"], linestyle='--', color=color, label=region)
plt.title("Rolling Average of Cancer Mortality Per Region")
plt.xlabel("Year")
plt.ylabel("Rolling Average Mortality")
plt.legend(fontsize=6, loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```
